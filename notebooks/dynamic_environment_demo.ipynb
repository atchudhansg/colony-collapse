## 🔍 Dynamic Resource Discovery

Find resources WITHOUT hardcoded positions!

```python
# Display all three levels
print(env.render_map(MapLevel.GROUND, use_emoji=True))
print(env.render_map(MapLevel.MOUNTAIN, use_emoji=True))
print(env.render_map(MapLevel.CAVE, use_emoji=True))
```

```python
def find_nearest_resources(env, sailor_id, resource_type=None, max_distance=50):
    """
    Dynamically find nearest resources of a given type.
    No hardcoded positions!
    
    Args:
        env: Environment
        sailor_id: Which sailor is searching
        resource_type: ResourceType or None for any
        max_distance: Search radius
    
    Returns:
        List of (resource, distance, resource_id) sorted by distance
    """
    sailor = env.state.sailors[sailor_id]
    results = []
    
    for res_id, resource in env.state.world_map.resources.items():
        if resource.gathered:
            continue
        
        # Filter by type if specified
        if resource_type and resource.resource_type != resource_type:
            continue
        
        # Only same level (for now - could expand to multi-level search)
        if resource.position.level != sailor.position.level:
            continue
        
        distance = resource.position.distance_to(sailor.position)
        if distance <= max_distance:
            results.append((resource, distance, res_id))
    
    return sorted(results, key=lambda x: x[1])

# Test it!
sailor_id = env.state.get_active_sailor()
wood = find_nearest_resources(env, sailor_id, ResourceType.WOOD, max_distance=100)

print(f"🌲 Found {len(wood)} wood resources for {sailor_id}:")
for res, dist, _ in wood[:5]:
    print(f"  {res.position.to_tuple()} - distance: {dist:.1f}")
```

## 🪜 Dynamic Stairs Finding

Find stairs WITHOUT hardcoded positions!

```python
def find_nearest_stairs(env, sailor_id, direction='UP'):
    """
    Dynamically find nearest stairs.
    
    Args:
        env: Environment
        sailor_id: Which sailor
        direction: 'UP' or 'DOWN'
    
    Returns:
        (stair_position, target_level, distance) or None
    """
    sailor = env.state.sailors[sailor_id]
    current_level = sailor.position.level
    
    stairs = []
    for pos1, pos2 in env.state.world_map.level_transitions:
        # Check if either end is on current level
        if pos1.level == current_level:
            goes_up = pos2.level.value > pos1.level.value
            if (direction == 'UP' and goes_up) or (direction == 'DOWN' and not goes_up):
                dist = pos1.distance_to(sailor.position)
                stairs.append((pos1, pos2.level, dist))
        
        elif pos2.level == current_level:
            goes_up = pos1.level.value > pos2.level.value
            if (direction == 'UP' and goes_up) or (direction == 'DOWN' and not goes_up):
                dist = pos2.distance_to(sailor.position)
                stairs.append((pos2, pos1.level, dist))
    
    if not stairs:
        return None
    
    return min(stairs, key=lambda x: x[2])

# Test it!
sailor_id = env.state.get_active_sailor()
stair = find_nearest_stairs(env, sailor_id, 'UP')

if stair:
    pos, target_level, dist = stair
    print(f"🪜 Nearest UP stairs for {sailor_id}:")
    print(f"   Location: {pos.to_tuple()}")
    print(f"   Goes to: {target_level.name}")
    print(f"   Distance: {dist:.1f}")
else:
    print("❌ No stairs found (might be on top/bottom level)")
```

## 🧭 Smart Navigation (No Hardcoding!)

Navigate to ANY position dynamically with obstacle avoidance.

```python
def navigate_to(env, sailor_id, target_pos, max_steps=50, verbose=False):
    """
    Navigate sailor to target position with obstacle avoidance.
    Works on ANY map - no hardcoded paths!
    
    Returns:
        (success, steps_taken, reason)
    """
    sailor = env.state.sailors[sailor_id]
    
    # Level mismatch check
    if sailor.position.level != target_pos.level:
        return (False, 0, "Different levels - use level transitions first")
    
    steps = 0
    stuck_count = 0
    
    while steps < max_steps:
        # Check whose turn it is
        active = env.state.get_active_sailor()
        if active != sailor_id:
            # Not our turn - skip
            action = Action(sailor_id=active, action_type=ActionType.WAIT)
            env.step({active: action})
            continue
        
        sailor = env.state.sailors[sailor_id]  # Refresh
        
        # Check if arrived
        if sailor.position == target_pos:
            if verbose:
                print(f"✅ Arrived in {steps} steps")
            return (True, steps, "Arrived")
        
        # Calculate direction
        dx = target_pos.x - sailor.position.x
        dy = target_pos.y - sailor.position.y
        
        # Determine primary direction
        if abs(dx) > abs(dy) and dx != 0:
            primary = ActionType.MOVE_EAST if dx > 0 else ActionType.MOVE_WEST
            secondary = (ActionType.MOVE_SOUTH if dy > 0 else ActionType.MOVE_NORTH) if dy != 0 else None
        elif dy != 0:
            primary = ActionType.MOVE_SOUTH if dy > 0 else ActionType.MOVE_NORTH
            secondary = (ActionType.MOVE_EAST if dx > 0 else ActionType.MOVE_WEST) if dx != 0 else None
        else:
            break
        
        # Try primary direction
        action = Action(sailor_id=sailor_id, action_type=primary)
        obs, _, _, _, info = env.step({sailor_id: action})
        
        if info.get(sailor_id, {}).get('success'):
            steps += 1
            stuck_count = 0
            if verbose and steps % 10 == 0:
                print(f"  Step {steps}: {sailor.position.to_tuple()}")
            continue
        
        # Try secondary if primary blocked
        if secondary:
            active = env.state.get_active_sailor()
            if active == sailor_id:
                action = Action(sailor_id=sailor_id, action_type=secondary)
                obs, _, _, _, info = env.step({sailor_id: action})
                if info.get(sailor_id, {}).get('success'):
                    steps += 1
                    stuck_count = 0
                    continue
        
        # Both blocked - try all 4 directions
        stuck_count += 1
        if stuck_count > 3:
            return (False, steps, f"Stuck at {sailor.position.to_tuple()}")
        
        for try_dir in [ActionType.MOVE_NORTH, ActionType.MOVE_SOUTH, 
                       ActionType.MOVE_EAST, ActionType.MOVE_WEST]:
            active = env.state.get_active_sailor()
            if active != sailor_id:
                break  # Lost our turn
            
            action = Action(sailor_id=sailor_id, action_type=try_dir)
            obs, _, _, _, info = env.step({sailor_id: action})
            if info.get(sailor_id, {}).get('success'):
                steps += 1
                stuck_count = 0
                break
    
    return (False, steps, "Max steps reached")

# Test navigation to nearest wood
sailor_id = env.state.get_active_sailor()
wood_resources = find_nearest_resources(env, sailor_id, ResourceType.WOOD)

if wood_resources:
    target = wood_resources[0][0]  # Nearest wood
    print(f"🎯 Navigating {sailor_id} to wood at {target.position.to_tuple()}")
    
    success, steps, reason = navigate_to(env, sailor_id, target.position, verbose=True)
    print(f"\nResult: {reason} ({steps} steps)")
else:
    print("No wood found on current level")
```

## 🔄 Multi-Episode Training Demo

Show how different seeds create different scenarios.

```python
def run_episode(seed=None, max_turns=50):
    """
    Run one complete episode with dynamic environment.
    Each episode has:
    - Different map layout (if seed=None)
    - Different traitor
    - Different resource positions
    """
    env = MaroonedEnv(seed=seed)
    obs = env.reset()
    
    episode_data = {
        'seed': env.state.seed,
        'traitor': env.state.traitor_id,
        'resources_found': 0,
        'resources_gathered': 0,
        'total_steps': 0,
    }
    
    for turn in range(max_turns):
        active = env.state.get_active_sailor()
        sailor = env.state.sailors[active]
        
        # Simple strategy: find and gather nearest resource
        nearby = find_nearest_resources(env, active, max_distance=20)
        
        if nearby:
            episode_data['resources_found'] += 1
            target, dist, res_id = nearby[0]
            
            # Navigate to it
            if sailor.position == target.position:
                # Gather
                action = Action(
                    sailor_id=active,
                    action_type=ActionType.GATHER_RESOURCE,
                    target_resource_id=res_id
                )
                obs, _, _, _, info = env.step({active: action})
                if info.get(active, {}).get('success'):
                    episode_data['resources_gathered'] += 1
            else:
                # Move towards resource
                dx = target.position.x - sailor.position.x
                dy = target.position.y - sailor.position.y
                
                if abs(dx) > abs(dy) and dx != 0:
                    move = ActionType.MOVE_EAST if dx > 0 else ActionType.MOVE_WEST
                elif dy != 0:
                    move = ActionType.MOVE_SOUTH if dy > 0 else ActionType.MOVE_NORTH
                else:
                    move = ActionType.WAIT
                
                action = Action(sailor_id=active, action_type=move)
                env.step({active: action})
        else:
            # Random exploration
            move = random.choice([ActionType.MOVE_NORTH, ActionType.MOVE_SOUTH,
                                ActionType.MOVE_EAST, ActionType.MOVE_WEST])
            action = Action(sailor_id=active, action_type=move)
            env.step({active: action})
        
        episode_data['total_steps'] += 1
    
    return episode_data

# Run 5 episodes with different random seeds
print("🎮 Running 5 episodes with DIFFERENT maps/roles:\n")
for i in range(5):
    result = run_episode(seed=None, max_turns=30)
    print(f"Episode {i+1}:")
    print(f"  Seed: {result['seed']}")
    print(f"  Traitor: {result['traitor']}")
    print(f"  Resources found: {result['resources_found']}")
    print(f"  Resources gathered: {result['resources_gathered']}")
    print()
```
